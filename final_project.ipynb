### dataset1: Video game sales

The video game sales dataset contains 16,598 games with sales greater than 100,000 copies, generated by a scrape of vgchartz.com . It has 11 attributes including: 
* Rank-ranking of overall sales
* Name-the video game name
* Platform-platform of the games release(i.e. Wii, Nes, GB, PC, PS4, etc. 31 values)
* Year-year of the game’s release
* Genre-genre of the game(i.e. action, sports, etc. 12 values)
* Publisher-publisher of the game(i.e. electronic arts, Activision, Nintendo, etc. 579 values)
* NA_sales-total sales in North America(in million)
* EU_sales-total sales in Europe(in million)
* JP_sales-total sales in Japan(in million)
* Other_sales-total sales in the rest of the world(in million)
* Global_sales-total worldwide sales

We think it’s an interesting dataset to work on, we can get several hypothesis from it, like
* Is the ranking associates with genre or the platform?
* What is the trend of genre of popular video games over years? Does it have relationship with the platform?  How does it distributed?
* What is the most popular platform for video games over the years? We can analyze it with the platform publication news.
* Does different publisher has a trend to publish a specific genre of video games?
* Compare the sales over different areas, compute the ratio between sales and population, does it have relationship with the development of the area?
* Does different areas has different favor of genre or platform of the video game? We can analyze it considering the culture and publisher of areas. 

We can visualize the above questions with kinds of vivid plots, and tidy the data focusing on each individual question.
